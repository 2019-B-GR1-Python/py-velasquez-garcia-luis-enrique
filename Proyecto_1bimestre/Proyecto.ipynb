#### Tema: Estadisticas de Asesinatos en Ecuador, año 2014
##### En 2014 Ecuador sufrió los peores años de violencia criminal, con más de 1,200 asesinatos y homicidios . El gobierno implementó varias medidas que han logrado reducir el número de víctimas a 1,010 en 2015: una disminución del 60% en cinco años. Si ponemos esta cifra en proporción con la población, vemos que el número de asesinatos y homicidios por cada 100 mil habitantes (la métrica más usada a nivel internacional), se redujo de 18,7 en 2010 a 6,4 en 2015, muy cerca del valor fijado por las Naciones Unidas como límite de lo que se considerara un país seguro, 5 casos por cada 100 mil habitantes. Es difícil hablar de seguridad desde las frías cifras: detrás de cada número hay una historia trágica. Sin embargo, los datos son importantes para saber si las estrategias implementadas dan resultados.



```python
import pandas as pd
import numpy as np
import math
import matplotlib
import matplotlib.pyplot as plt
```

```python
path = 'asesinatos2014.xlsx'
```

```python
df = pd.read_excel(path)
df
```

```python
df.groupby('Provincia').mean()["Edad"].plot(kind='bar',stacked=True,title="Promedio de edad de asesinatos en cada provincia")
```

```python
df.groupby('tipo_muert_matriz').mean()["Edad"].plot(kind='bar',stacked=True,title="Promedio de edad por tipo de muerte")
```

```python
# Cantidad de muertes 
tipo_muerte_edad = df.groupby(['hora'])['Edad'].describe()
tipo_muerte_edad.plot(kind='line')
```

```python
df.groupby('Sexo').count()["Edad"].plot(kind='bar',stacked=True,title="Numero de muerte comparados entre hombres mujeres")
```

```python
df.groupby('Estado Civil').count()["Edad"].plot(kind='bar',stacked=True,title=" Cantidad de muertes por estado civil")
```

```python
df.groupby('Provincia').count()['Edad'].plot(kind='bar',stacked=True,title="cantidad de muertes por provincia")
```

```python
df.groupby('tipo_muert_matriz').count()["Edad"].plot(kind='bar',stacked=True,title="cantidad de tipos de muerte")
```

```python
fig, ax1 = plt.subplots()
asesinato = pd.read_excel("asesinatos2014.xlsx")
my_plot = asesinato.plot("Edad", "hora", kind="scatter", ax=ax1)

ax1.set_xlabel("Edades")
ax1.tick_params(labelsize=16, pad=8)
fig.suptitle('Diagrama de dispersión de las horas y edades en las cuales ocurre una muerte', fontsize=15)
```

```python
# Agrupamos por edad y sexo, y se calcula el total
provincia_y_sexo = df.groupby(['Provincia','Sexo'])
total = provincia_y_sexo['Edad'].count()

provincia_y_sexo = df.groupby(['Provincia','Sexo'])
total = provincia_y_sexo['Edad'].count()
total.unstack()

# Grafico de la cantidad para cada sexo es apilado por sitio.
provincia_y_sexo = df.groupby(['Provincia','Sexo'])
total = provincia_y_sexo['Edad'].count()
edad_sexo = total.unstack()
s_plot = edad_sexo.plot(kind='bar',stacked=True,title="Cantidad de muertes por provincia diferenciando los sexos")
s_plot.set_ylabel("Edad")
s_plot.set_xlabel("Provincia")
```

```python
# cantidad de muertes por mes
conteo_horas = df.groupby('mes')['Edad'].count()
conteo_horas.plot(kind='line')
```

```python
df.groupby('mes').count()["Edad"].plot(kind='bar',stacked=True,title="concurrencia de muertes por mes")
```
